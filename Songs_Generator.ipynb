```python
# code to download binary with photon counts

song = 'song.txt'
data_status = os.path.exists('./' + song)

if data_status == False:
    url = 'https://drive.google.com/uc?export=view&id=16BOKCzifjMjF5A_xr4jGT3UYPSoOc4jR'
    wget.download(url, song)
```

```python
vocab = dict()
text = list()
with open(song, "r") as songwords:
    for line in songwords:
      s = line.split()
      for i in s:
        text.append(i)

for i in range(len(text)-1):
  y = [(text[i], dict())]
  vocab.update(y)

for i in range(len(text)-1):
  s1 = text[i]
  s2 = text[i+1]
  count = 0
  if s2 not in vocab[text[i]]:
    for j in range(len(text)-1):
      if text[j] == s1 and text[j+1] == s2:
        count += 1
    y = [(text[i+1], count)]
    vocab[text[i]].update(y)

words = list(vocab.keys())

print(vocab)
print(words)
```

```python
n = len(vocab)
prob_matrix = np.zeros((n, n))

for i in range(n):
  keys = list(vocab[words[i]].keys())
  values = list(vocab[words[i]].values())
  l = len(keys)
  for j in range(l):
    for h in range(n):
      if keys[j] == words[h]:
        prob_matrix[i][h] = vocab[words[i]][keys[j]] / sum(values)

print(prob_matrix)
```

```python
dot_len = 16
dot_count = 0

first_words = [word for word in vocab if word[0] == word[0].upper()]

word = rm.choice(first_words)
print(word, end = ' ')
if word[-1] in '.?!':
    print()
    dot_count += 1

while dot_count < dot_len:
  index = words.index(word)
  word = np.random.choice(words, replace = True, p = prob_matrix[index])
  print(word, end=' ')
  if word[-1] in '.?!':
    print()
    dot_count += 1
```
